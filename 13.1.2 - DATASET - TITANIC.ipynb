```python
#Criando uma funo que fara isso para todos os nomes
```

```python
def tratar_nome(linha):
    import re
    match=re.compile(r'"(.*)(,)(\s.*)"')
    return match.sub (r'"\3 \1"',linha)
```

# Tirando a Media da idade dos pasageiros do TITANIC

```python
#PRIMEIRO PRECISAMOS SUBSTITUIR OS ESPAOS VAZIOS EM IDADE E TRANFORMAR OS DADOS DE IDADE EM UM NUMERO FLOAT
for indice, age in enumerate(titanic_dados['Age']):
    if titanic_dados['Age'][indice]!= '':
        titanic_dados['Age'][indice]= float(age)
    else:
        titanic_dados['Age'][indice]=-1
            
    
```

```python
#AGORA VAMOS COPIAR PARA UMA LISTA SOMENTE AS IDADES QUE NAO FORAM SUBSTITUIDAS POR -1 PARA TIRARMOS  A media
lista_idade=[]
for idade in titanic_dados['Age']:
    if idade != -1:
        lista_idade.append(idade)
```

```python
def media(lista): #definindo uma funo que tira a media de idade do banco de dados
    soma_total=sum(lista)
    quantidade=len(lista)
    return soma_total/quantidade
```

```python
media_idade= media(lista_idade)
```
